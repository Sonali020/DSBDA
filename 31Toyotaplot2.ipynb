31.	Plot following data visualization using Python matplotlib library by plotting histogram, scatter-plot and bar-plot (Toyota.csv)
a.	Age vs Price- Scatter plot
b.	Distribution of KM data - histogram
c.	FuelType wise CarsCount- Barplot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Toyota.csv")
```

```python
df
```

#  a. Age vs Price – Scatter Plot

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Price'], color='green', edgecolor='black')
plt.title("Scatter Plot: Age vs Price")
plt.xlabel("Age (months)")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

#  b. Distribution of KM – Histogram

```python
plt.figure(figsize=(8, 6))
plt.hist(df['KM'], bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram: Distribution of Kilometers Driven")
plt.xlabel("Kilometers (KM)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

# c. FuelType-wise Car Count – Bar Plot

```python
plt.figure(figsize=(8, 6))

# Count of each fuel type
fuel_counts = df['FuelType'].value_counts()

# Plot bar chart
plt.bar(fuel_counts.index, fuel_counts.values, color='orange', edgecolor='black')
plt.title("Bar Plot: Fuel Type Wise Car Count")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```
