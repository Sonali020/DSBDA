11. Perform the following operations in python on Titanic-Dataset.csv Dataset
    a.	Handle missing Age and Cabin values using appropriate imputation techniques.
    b.	Convert Sex and Embarked columns into numeric form using encoding.
    c.	Create a new feature FamilySize = SibSp + Parch.
    d.	Bin Fare into price categories (e.g., Low, Medium, High).


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
```

```python
df
```

# a. Handle missing Age and Cabin values

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Cabin'] = df['Cabin'].fillna('Unknown')
```

# b. Convert Sex and Embarked to numeric form using encoding

```python
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})
```

```python
df['Embarked'] = df['Embarked'].fillna('S')  # Fill with most common port
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

# c. Create a new feature FamilySize = SibSp + Parch

```python
df['FamilySize'] = df['SibSp'] + df['Parch']
```

# d. Bin Fare into categories: Low, Medium, High

```python
df['FareCategory'] = pd.cut(df['Fare'],
                            bins=[-1, 30, 100, np.inf],
                            labels=['Low', 'Medium', 'High'])
```
