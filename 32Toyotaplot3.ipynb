32.	Visualize the data using Python Seaborn /matplotlib library by plotting histogram, scatter-plot and bar-plot (Toyota.csv)
a.	Age vs Price- Scatter plot
b.	Distribution of KM data - histogram
c.	FuelType wise CarsCount- Barplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Toyota.csv")
```

#  a. Scatter Plot â€“ Age vs Price

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Price', data=df, color='teal', edgecolor='black')
plt.title("Scatter Plot: Age vs Price")
plt.xlabel("Age of Car (in months)")
plt.ylabel("Price")
plt.grid(True)
plt.show()
```

#  b. Histogram â€“ Distribution of KM

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='KM', bins=30, kde=True, color='steelblue')
plt.title("Histogram: Kilometers Driven")
plt.xlabel("Kilometers (KM)")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

# c. Bar Plot â€“ FuelType-wise Car Count

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='FuelType', data=df, palette='pastel', edgecolor='black')
plt.title("Bar Plot: Count of Cars by Fuel Type")
plt.xlabel("Fuel Type")
plt.ylabel("Number of Cars")
plt.grid(True)
plt.show()
```
