1.	Perform the following operations in python on given dataset [Iris.csv]
a.	Subset rows where petal length > 1.5 and species is "setosa".
b.	Merge the Iris dataset with a custom species info table (e.g., color or habitat).
c.	Sort flowers by sepal_width and then by sepal_length.
d.	Transpose the dataset’s first 5 rows to view feature-wise comparisons.
e.	Reshape the dataset using melt() and pivot() to transform features into long and wide formats.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Iris.csv')
df
```

# a.	Subset rows where petal length > 1.5 and species is "setosa".

```python
subset=df[(df['PetalLengthCm']>1.5) & (df['Species']=='Iris-setosa')]
```

```python
subset
```

# b.	Merge the Iris dataset with a custom species info table (e.g., color or habitat).

```python
species_info = pd.DataFrame({
    'Species': ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
    'Color': ['Red', 'Blue', 'Purple'],
    'Habitat': ['Garden', 'Meadow', 'Swamp']
})
```

```python
merged_df = pd.merge(df, species_info, on='Species', how='left')
print(merged_df.head())
```

# c.	Sort flowers by sepal_width and then by sepal_length.

```python
sort = df.sort_values(by=['SepalWidthCm', 'SepalLengthCm'])
```

```python
sort
```

# d.	Transpose the dataset’s first 5 rows to view feature-wise comparisons.

```python
transposed_df=df.head().transpose()
```

```python
transposed_df
```

# e.	Reshape the dataset using melt() and pivot() to transform features into long and wide formats.

```python
# Melting (long format)
melted_df = pd.melt(df, id_vars=["Id", "Species"], 
                    value_vars=["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"],
                    var_name="Feature", value_name="Value")
print(melted_df.head())
```

```python
# Pivoting (wide format)
pivot_df = melted_df.pivot_table(index=["Id", "Species"], columns="Feature", values="Value").reset_index()
print(pivot_df.head())
```
