30.	Plot following using Python libraries matplotlib / seaborn (Toyota.csv)
a.	Scatter plot- Car-Price by Age
b.	Histogram on Cars data KM
c.	Bar plot on counts of FuelType category (Petrol, Disel and CNG)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Toyota.csv")
```

```python
print(df.head())
```

```python
print(df.info())
```

# a. Scatter Plot â€“ Car Price by Age

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Age', y='Price')
plt.title("Scatter Plot: Car Price by Age")
plt.xlabel("Age of Car (in months)")
plt.ylabel("Price of Car")
plt.grid(True)
plt.show()
```

# b. Histogram â€“ Cars Data KM

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='KM', bins=30, kde=True)
plt.title("Histogram: Kilometers Driven")
plt.xlabel("Kilometers (KM)")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

# c. Bar Plot â€“ Counts of FuelType (Petrol, Diesel, CNG)

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='FuelType', order=df['FuelType'].value_counts().index)
plt.title("Bar Plot: Count of FuelType")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```
